```python
import statistics as st
list = [1,2,3,4,5,6]
print(list)
mean = st.mean(list)
std = st.stdev(list)
for i in range(6):
    list[i] = (list[i] - mean) / std
print(list)
mean = st.mean(list)
std = st.stdev(list)
print(mean)
print(std)
```
